```python
def convertIdioticMatToDataFrame (mat, columnName):
    convertedData = []
    data = mat[columnName][0]
    for row in data:                               # 2048 row of data(training)
        timeSequence = [i[0] for i in data[0]]     # In every row of data, there is a time sequence data with length = 8192
        convertedData.append(timeSequence)
    return pd.DataFrame(np.array(convertedData))
```
