```python
# convert whaterver is in the .mat file in to dataframe
trainingDataframe = pd.DataFrame(np.reshape(mat_train_feature["data_train"][0], (2048, 1)))
trainingLabel = pd.DataFrame(np.reshape(mat_train_label["data_train_labels"][0], (2048, 1)))
trainingLabel
```

```python
# convert the data to one dimentional array
for i in range(len(trainingDataframe)):
    row = trainingDataframe.loc[i].values[0]
    dimReducedRow = [value[0] for value in row]
    trainingDataframe.at[i, 0] = dimReducedRow

#trainingDataframe.head()
```

```python
# add label and column name to dataframe
trainingDataframe["defect_type"] = mat_train_label["data_train_labels"][0]
trainingDataframe.columns = ["accelration", "defect_type"]
trainingDataframe.head()
```

```python
trainingDataframe.to_csv("../Dataset/data_train.csv", index = False)
```
